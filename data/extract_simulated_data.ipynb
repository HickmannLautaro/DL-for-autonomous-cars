```python
import numpy as np
import cv2
import glob

from matplotlib import pyplot as plt
```

```python
file_list=glob.glob("runs/**/discrete_histories.npz")
print(file_list)
```

```python
file_list = ['runs/test_trees/discrete_histories.npz']#[np.random.choice(file_list,1)[0]]
file_list
```

# Normal 3 channel data

```python
dataset='road_tree_continuous_test'
```

```python
for file in file_list:
    data = np.load(file)
    for i, image in enumerate(data.f.frames):
        image= cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        cv2.imwrite(f"styletransfer/cut/datasets/{dataset}/trainA/patch_{i}.jpg", image[0:256,0:256])
```

# binary mask 4 channel data

```python
for file in file_list:
    data = np.load(file)
    for i, image in enumerate(data.f.frames):
        #print(image.shape)
        aux2=np.zeros((256,256,1), dtype=np.uint8)
        aux2[np.where(image[:256, :, 1] == 105)]= [255]
        aux2[np.where(image[:256, :, 1] == 107)]= [255]
        aux2[np.where(image[:256, :, 1] == 102)]= [255]
        image= cv2.cvtColor(aux2, cv2.COLOR_RGB2BGR)
      #  image= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        cv2.imwrite(f"styletransfer/cut/datasets/{dataset}/segA/patch_{i}.jpg", image)
    break
```
